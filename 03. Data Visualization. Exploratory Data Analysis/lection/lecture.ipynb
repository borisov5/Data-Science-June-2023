```python
whr_merged = whr_merged.merge(whr_2017.rename(columns=lambda col: col + "_2017"), left_on = "Country or region", right_on= "Country_2017", suffixes=["", "_2017"])
```

```python
whr_merged = whr_merged.rename(columns = {"Happiness.Score_2017": "Score_2017"})
```

```python
scores = whr_merged[["Country or region", "Score_2017", "Score_2018", "Score_2019"]]
```

```python
bg_scores = scores[scores["Country or region"] == "Bulgaria"].T.drop("Country or region").values.ravel()
ro_scores = scores[scores["Country or region"] == "Romania"].T.drop("Country or region").values.ravel()
```

```python
years = list(range(2017, 2020))
```

```python
plt.scatter(years, bg_scores)
plt.xticks(years)
plt.show()
```

```python
plt.plot(years, bg_scores, linewidth = 2, linestyle = "dashed", label = "BG")
plt.scatter(years, bg_scores, marker="s", s = 50)

plt.plot(years, ro_scores, linewidth = 2, linestyle = "dashed", label = "RO")
plt.scatter(years, ro_scores, marker="s", s = 50)

plt.title("Bulgaria: Happiness Score")
plt.xlabel("Year")
plt.ylabel("Score")
plt.xticks(years)
plt.legend(loc = "lower right")
plt.show()

# plt.savefig("figure.jpeg") # svg, pdf, eps
```

```python
g = nx.florentine_families_graph()
pos = nx.spring_layout(g)
nx.draw_networkx(g)

print(nx.shortest_path(g, "Pazzi", "Peruzzi"))
print(nx.shortest_path_length(g, "Pazzi", "Peruzzi") +1 )
```

```python
centrality = nx.betweenness_centrality(g)
print(centrality)

max_value = max(centrality.values())
print(max_value)
value = {i for i in centrality if centrality[i]==max_value}
print(value)
```
